**Generators vs Return, which would benefit memory efficiency, why should we prefer one over the other while process big data**<br>
Python Generators are functions that outputs/returns an iterable generator object. Generator objects are fetched one at a time. A Generator function is similar to a normal function; however, it does not include a return value and instead uses a yield statement.<br><br>
The main difference between a generator function (yield statement) and a normal function (return statement), a generator function instead returns a generator object which has the value we had given to the yield statement. However, we go not get a message/value outputted. The values from the generator can be read using for-in, list() and next() method.<br><br>
Another major difference between a yield statement and return statement is that after a return statement is executed the function stops running and the value is returned to the caller. Return statement causes a function to exit. It terminates the execution of a function and destroys all the local variables.  But in case of generator function once the execution starts when it gets the first yield it stops the execution and gives back the generator object. You can use the generator object to get the values and, pause and resume back as per your requirement.<br><br>
**What are the benefits of Yield vs Return statement?**<br>
Yield statements are best when the data size is large, when the user needs to execute the function quickly on a large data set. Also, it is best to use yield when you want to return multiple big set of values to the calling function. No memory is used or taken up when using a yield statement.<br>
Compared to a return statement which allocates memory for the returned value. Therefore, if return statements are used on larger data sets, the use of memory will cause the performance or the execution of the code to drop significantly, this means the execution time will be longer as there is more to process.<br>
Therefore, it is best to use Return statements on smaller datasets and Yield statements on larger data sets.

**What is the use of Exploratory Data Analysis (EDA)?**<br>
EDA is an abbreviation for Exploratory Data Analysis. Exploratory Data Analysis is a process of investigation on the data set that we are trying to exact clear and to the point information and insights. This information can include testing hypothesises, and assumptions or identifying patterns as well as anomalies. We can gather this information by using statistics summaries and graphical representations. This is a crucial process to help us understand the data set we are working with before we try and gather insights.<br><br>
One of the first steps in understanding the data set is getting details such as number of rows, number of columns as well as their corresponding names and data types. This is done using data.info() function.<br><br>
The next information that needs to be recorded or considered is each columns minimum and maximum values, median values, their standard deviation as well as their mean value. This can be achieved by using the data.describe() function.<br><br>
The next step in the EDA process is checking for null entries and if needed we can fill in the null values with the median, mean or a numerical value. We are able to achieve it by using the data.IsNull() function. We are also able to check for duplicate values using data.duplicated() and if needed we are able to remove the duplicates from the data set.<br><br>
The last step of EDA is to then represent the data graphically.<br><br>
Once we understand the data set we can extract some clear and to the point information and insights. This is the reason why Exploratory Data Analysis is so critical.

```python
# ECDF
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data) 

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# Compute ECDFs
x_c1, y_c1 = ecdf(class1_alcohol)
x_c2, y_c2 = ecdf(class2_alcohol)
x_c3, y_c3 = ecdf(class3_alcohol)

# Plot all ECDFs on the same plot
plt.plot(x_c1, y_c1, marker='.', linestyle='none')
plt.plot(x_c2, y_c2, marker='.', linestyle='none')
plt.plot(x_c3, y_c3, marker='.', linestyle='none')

# Annotate the plot
plt.legend(('Class 1', 'Class 2', 'Class 3'), loc='lower right')
_ = plt.xlabel('Alcohol Percentage %')
_ = plt.ylabel('ECDF')

# Display the plot
plt.title('Plot to compare ECDFS to Alcohol Percentage for each Wine Class')
plt.show()

# Compute ECDFs
x_c1, y_c1 = ecdf(class1_color_intensity)
x_c2, y_c2 = ecdf(class2_color_intensity)
x_c3, y_c3 = ecdf(class3_color_intensity)

# Plot all ECDFs on the same plot
plt.plot(x_c1, y_c1, marker='.', linestyle='none')
plt.plot(x_c2, y_c2, marker='.', linestyle='none')
plt.plot(x_c3, y_c3, marker='.', linestyle='none')

# Annotate the plot
plt.legend(('Class 1', 'Class 2', 'Class 3'), loc='lower right')
_ = plt.xlabel('Colour Intensity')
_ = plt.ylabel('ECDF')

# Display the plot
plt.title('Plot to compare ECDFS to Colour Intensity for each Wine Class')
plt.show()

# Compute ECDFs
x_c1, y_c1 = ecdf(class1_flavanoids)
x_c2, y_c2 = ecdf(class2_flavanoids)
x_c3, y_c3 = ecdf(class3_flavanoids)

# Plot all ECDFs on the same plot
plt.plot(x_c1, y_c1, marker='.', linestyle='none')
plt.plot(x_c2, y_c2, marker='.', linestyle='none')
plt.plot(x_c3, y_c3, marker='.', linestyle='none')

# Annotate the plot
plt.legend(('Class 1', 'Class 2', 'Class 3'), loc='lower right')
_ = plt.xlabel('Flavanoids')
_ = plt.ylabel('ECDF')

# Display the plot
plt.title('Plot to compare ECDFS to Flavanoids for each Wine Class')
plt.show()

# Compute ECDFs
x_c1, y_c1 = ecdf(class1_malic_acid)
x_c2, y_c2 = ecdf(class2_malic_acid)
x_c3, y_c3 = ecdf(class3_malic_acid)

# Plot all ECDFs on the same plot
plt.plot(x_c1, y_c1, marker='.', linestyle='none')
plt.plot(x_c2, y_c2, marker='.', linestyle='none')
plt.plot(x_c3, y_c3, marker='.', linestyle='none')

# Annotate the plot
plt.legend(('Class 1', 'Class 2', 'Class 3'), loc='lower right')
_ = plt.xlabel('Malic Acid')
_ = plt.ylabel('ECDF')

# Display the plot
plt.title('Plot to compare ECDFS to Malic Acid for each Wine Class')
plt.show()
```
